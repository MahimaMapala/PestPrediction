```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df[numeric_cols].hist(bins=30, figsize=(20, 15))
plt.show()
```

```python
# finding features that matter
from sklearn.feature_selection import mutual_info_classif



X = df[numeric_cols]
y = df['disease_name']

mi = mutual_info_classif(X, y, discrete_features=False)
mi_series = pd.Series(mi, index=numeric_cols).sort_values(ascending=False)

mi_series.plot(kind='bar', color='skyblue')
plt.title('Mutual Information of Numeric Features')
plt.ylabel('Mutual Information')
plt.xlabel('Feature')
plt.tight_layout()
plt.show()
```

```python
#finding redundant features
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = pd.DataFrame()
vif_data["feature"] = numeric_cols
vif_data["VIF"] = [variance_inflation_factor(df[numeric_cols].values, i) for i in range(len(numeric_cols))]
print(vif_data)

sns.barplot(x="VIF", y="feature", data=vif_data, palette="viridis")
plt.show()
```

```python
#feature correlation
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.show()

#can fill leaf and precipitation with humid min
```

```python
from scipy.stats import pointbiserialr

correlations = pd.DataFrame(index=numeric_cols, columns=df['disease_name'].unique())

for disease in df['disease_name'].unique():
    y_bin = (df['disease_name'] == disease).astype(int)
    for feature in numeric_cols:
        corr, _ = pointbiserialr(df[feature], y_bin)
        correlations.loc[feature, disease] = corr

correlations = correlations.astype(float)

plt.figure(figsize=(12, 6))
sns.heatmap(correlations, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('Point Biserial Correlation - numeric_cols vs disease_name')
plt.show()
```

```python
missing_by_class = df.groupby('disease_name')[['Leaf wetness probability_mean', 'Precipitation Total_mean']].apply(lambda x: (x == 0).sum())

missing_by_class.plot(kind='bar', figsize=(10, 6))
plt.title('Count of Zero Values by Disease Class')
plt.ylabel('Count of Zero Values')
plt.xlabel('Disease Name')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
